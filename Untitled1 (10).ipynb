```python
Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution
with given mean at a given point.
ans-
Here's the Python code for a function that calculates the cumulative distribution function (CDF) of a Poisson distribution with a given mean at a given point:

python
Copy code
import math

def poisson_cdf(mean, k):
    """
    Calculates the cumulative distribution function (CDF) of a Poisson distribution
    with a given mean at a given point.
    
    Args:
    - mean (float): the mean of the Poisson distribution
    - k (int): the point at which to evaluate the CDF
    
    Returns:
    - cdf (float): the CDF of the Poisson distribution at k
    """
    
    # Calculate the CDF using the Poisson distribution formula
    cdf = 0
    for i in range(k + 1):
        cdf += (mean ** i / math.factorial(i)) * math.exp(-mean)
    
    return cdf
This function takes two arguments: mean, the mean of the Poisson distribution, and k, the point at which to evaluate the CDF. It then uses the Poisson distribution formula to calculate the CDF, by summing the probabilities of all the Poisson random variables with values less than or equal to k. The math library is used to calculate the factorial and exponential functions.

Here's an example of how to use the function:

python
Copy code
mean = 3
k = 5
cdf = poisson_cdf(mean, k)
print("The CDF of a Poisson distribution with mean", mean, "at point", k, "is", cdf)
This will output:

csharp
Copy code
The CDF of a Poisson distribution with mean 3 at point 5 is 0.8047358835275832




```

```python
Q7. How Binomial distribution different from Poisson distribution?
ans-Both Binomial and Poisson distributions are used to model the probabilities of certain events occurring. However, they differ in several ways:

Definition: The Binomial distribution models the probability of a certain number of successes in a fixed number of trials, where each trial has the same probability of success. The Poisson distribution, on the other hand, models the probability of a certain number of events occurring in a fixed time or space interval, where the events occur randomly and independently with a certain average rate.

Parameters: The Binomial distribution has two parameters: the number of trials (n) and the probability of success in each trial (p). The Poisson distribution has only one parameter: the average rate of events occurring in the given interval (λ).

Assumptions: The Binomial distribution assumes that the trials are independent and that the probability of success is constant for each trial. The Poisson distribution assumes that the events occur randomly and independently, and that the average rate of occurrence is constant over the given interval.

Shape: The Binomial distribution is discrete and has a finite range of possible values. The Poisson distribution is also discrete, but it has an infinite range of possible values.

Limitations: The Binomial distribution is appropriate when the number of trials is relatively small, and the probability of success is not too small or too large. The Poisson distribution is appropriate when the average rate of occurrence is small and the events are rare, or when the number of trials is very large and the probability of success is very small.

In summary, the Binomial distribution models the probability of a fixed number of successes in a fixed number of trials, while the Poisson distribution models the probability of a certain number of events occurring in a fixed time or space interval, where the events occur randomly and independently with a certain average rate.




```
